```
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

np.random.seed(42)
sns.set(font_scale = 1.3)
```

```
#n_samples = 700
raw_data = make_moons(n_samples = 2000, noise = 0.25, random_state = 42)
```

```
data = raw_data[0]
target = raw_data[1]

print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
```

```
df = pd.DataFrame(data = np.c_[data, target], columns = ['x1', 'x2', 'target'])
df.head()
```

```
plt.figure(figsize=(8, 6))
plt.scatter(x = df['x1'], y = df['x2'], c = target, cmap = 'viridis')
plt.title('Klasyfikacja - dane do modelu')
plt.show()
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.3)

print(f'x_train shape: {x_train.shape}')
print(f'x_test shape: {x_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```
plt.figure(figsize=(8, 6))
plt.scatter(x = x_train[:,0], y = x_train[:,1], c = y_train, cmap = 'RdYlBu', label = 'Training set')
plt.scatter(x = x_test[:,0], y = x_test[:,1], c = y_test, cmap = 'RdYlBu', label = 'Test set', marker = 'x')
plt.title('Klasyfikacja - dane do modelu')
plt.legend()
plt.show()
```

```
from sklearn.linear_model import LogisticRegression
lin = LogisticRegression()

lin.fit(x_train, y_train)
print(f'Trainig set score: {lin.score(x_train, y_train)}')
print(f'Test set score: {lin.score(x_test, y_test)}')
```

```
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize = (8, 6))
plot_decision_regions(x_train, y_train, lin)
plt.title(f'Zbir treningowy: dokadno {lin.score(x_train, y_train):.4f}')
plt.show()
```

```
plt.figure(figsize = (8, 6))
plot_decision_regions(x_train, y_train, lin)
plt.title(f'Zbir testowy: dokadno {lin.score(x_test, y_test):.4f}')
plt.show()
```

```
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth = 1)
tree.fit(x_train, y_train)

plt.figure(figsize = (8, 6))
plot_decision_regions(x_train, y_train, tree)
plt.title(f'Zbir treningowy DecisionTreeClassifier, max_depth = 1: dokadno {tree.score(x_train, y_train):.4f}')
plt.show()
```

```
plt.figure(figsize = (8, 6))
plot_decision_regions(x_test, y_test, tree)
plt.title(f'Zbir testowy DecisionTreeClassifier, max_depth = 1: dokadno {tree.score(x_test, y_test):.4f}')
plt.show()
```

```
tree = DecisionTreeClassifier(max_depth = 6)
tree.fit(x_train, y_train)

plt.figure(figsize = (8, 6))
plot_decision_regions(x_train, y_train, tree)
plt.title(f'Zbir treningowy DecisionTreeClassifier, max_depth = 6: dokadno {tree.score(x_train, y_train):.4f}')
plt.show()

```

```
plt.figure(figsize = (8, 6))
plot_decision_regions(x_test, y_test, tree)
plt.title(f'Zbir testowy DecisionTreeClassifier, max_depth = 6: dokadno {tree.score(x_test, y_test):.4f}')
plt.show()
```
