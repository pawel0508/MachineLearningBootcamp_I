### Spadek wzdłóż gradientu

```
import pandas as pd
import numpy as np 
import plotly.express as px
```

### Generowanie danych

```

X1 = np.array([1, 2, 3, 4, 5, 6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)

print(f'Lata pracy: {X1}')
print(f'Wynagrodzenie: {Y}')
print(f'Liczba próbek: {m}')
```

```
X1 = X1.reshape(m, 1)
Y = Y.reshape(-1, 1)
bias = np.ones((m, 1))
```

```
X = np.append(bias, X1, axis = 1)
```

```
print(X)
```

### Losowa inicjalizacja wag

```
eta = 0.01
weights = np.random.randn(2, 1) # w liscie dwie listy jednoelementowe
print(weights) 
```

```
intercept = []
coef = []
```

```
for i in range(3000):
  gradient = (2/m) * X.T.dot(X.dot(weights) - Y)
  weights = weights - eta*gradient
  intercept.append(weights[0][0])
  coef.append(weights[1][0])
print(intercept[-1])
print(coef[-1])
print(weights)
```

```
df = pd.DataFrame(data = {'Intercept' : intercept, 'Coef' : coef})
```

```
df
```

```
px.line(data_frame=df, y = 'Intercept', width = 800, title = 'Dopasowanie Intercept')
```

```
px.line(data_frame=df, y = 'Coef', width = 800, title = 'Dopasowanie Coef')
```
